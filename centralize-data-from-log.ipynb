```python
# take raw.txt FS log text and clean it up to have one row per line with only file path, save to clean.txt

import re

lines = [] # Declare an empty list named "lines"
with open ('raw.txt', 'rt') as in_file: # Open file raw.txt for reading of text data.
    for line in in_file: 
        if re.search(r"L:", line) :
            line = line.rpartition('Send ')[-1]
            line = line.rpartition(' to dicom')[0]
            lines.append(line) # add that line to our list of lines. 

with open('clean.txt', 'w') as f:
    for line in lines:
        f.write(line + '\n')
        
```

```python
# analyze log file, extract path of files that had problems, copy just those files to new directory

import re, os
from shutil import copy2

# destpaths = []   
# sources = []
# dests = []

with open ('raw.txt', 'rt') as in_file: # Open file raw.txt for reading of text data.
    for line in in_file: 
        if re.search(r"L:", line) :  # searching for lines that contain relevant text
            result = line.rpartition('Send ')[-1]  # cutting out junk before relevant text
            source = result.rpartition(' to dicom')[0]  # cutting out more junk after relevant text leaving copy source path
            # sources.append(source)   
            dest = source.rpartition('TCGA-BLCA\\')[-1]  # cutting out relevant path info for copy destination
            dest = 'resubmit\\' + str(dest) # appending base dir for copy destination
            # dests.append(dest)
            destpath = (dest[0:-10]) # cutting off file name at the end to leave only the path for directory creation
            # destpaths.append(destpath)
            if not os.path.exists(destpath):
                os.makedirs(destpath)
            copy2(source,dest)

# print(sources[0])
# print(dests[0])
# print(destpaths[0])
```
